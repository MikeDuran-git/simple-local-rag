# Créer un menu

Afin d'etre sur que nous traitons chaque cas, il faut savoir en avance combien de repas compte un menu. On sait que dans les restaurants des menu sont composées: 'appetizer', 'dinner', 'lunch', 'breakfast_and_brunch', 'desert'.



```python
def create_menu(ingredient_list,model='ft:gpt-3.5-turbo-1106:personal:recipecreatorv7b:9XU88CKN'):
    menu = []
    # we count the number of ingredient list to create the menu
    for key in ingredient_list:
        recipe = generate_recipe(
            ingredients=ingredient_list[key],
            recipe_type=key,
            model=model
        )
        menu.append(recipe)

    return menu
```

```python
# we test our method with the following ingredient list
ingredient_list = {
    "breakfast_and_brunch": 'oatmeal, almond milk, banana, Whey protein cookies and cream, honey',
    "appetizer": 'banana, apple',
    "lunch": 'skinless chicken breast, rice, broccoli',
    "desert": 'rice cake, rice cake, protein powder, banana',
    "dinner": 'raw salmon, brown rice, broccoli',
}

menu = create_menu(ingredient_list,model='gpt-4o-mini')
for recipe in menu:
    print(print_recipe(recipe))
```

```python
def create_menu(number_of_meals_per_day_list, days, ingredient_list,model='ft:gpt-3.5-turbo-1106:personal:recipecreatorv7b:9XU88CKN'):
    menu = {}
    # Mapping for meal types based on number of meals per day
    meal_types = {
        1: ["dinner"],
        2: ["lunch", "dinner"],
        3: ["breakfast_and_brunch", "lunch", "dinner"],
        4: ["breakfast_and_brunch", "lunch", "dessert", "dinner"],
        5: ["breakfast_and_brunch", "appetizer", "lunch", "dessert", "dinner"],
    }

    for day, number_of_meals_per_day in zip(days, number_of_meals_per_day_list):
        daily_meals = {}
        meals_per_day = meal_types[number_of_meals_per_day]
        for meal_type in meals_per_day:
            ingredients = ingredient_list.get(meal_type, "")
            recipe = generate_recipe(ingredients=ingredients, recipe_type=meal_type,diet="vegan",model=model)
            daily_meals[meal_type] = recipe
        menu[day] = daily_meals

    return menu

# Test the create_menu function
if __name__ == "__main__":
    # Example usage

    # Example list of number of meals per day for each day
    number_of_meals_per_day_list = [3, 2, 4]  # For Monday, Tuesday, and Wednesday
    days=["monday", "tuesday", "wednesday"]
    # Since we anticipate that the user eats the same meal every now and then we can create a menu for the week
    ingredient_list = {
        "breakfast_and_brunch": 'oatmeal, almond milk, banana, Whey protein cookies and cream, honey',
        "appetizer": 'banana, apple',
        "lunch": 'skinless chicken breast, brown rice, broccoli',
        "dessert": 'rice cake, rice cake, protein powder, banana',
        "dinner": 'raw salmon, brown rice, broccoli',
    }

    menu = create_menu(number_of_meals_per_day_list,days, ingredient_list,model='gpt-4o-mini')
```

```python
import re

def extract_ingredients(recipe):
    """Extracts and returns a list of ingredients from a recipe string."""
    pattern = re.compile(r'<ingredient_start>(.*?)<ingredient_end>', re.DOTALL)
    match = pattern.search(recipe)
    if match:
        ingredients_section = match.group(1).strip()
        ingredients = ingredients_section.split('<ingredient_next>')
        return [ingredient.strip() for ingredient in ingredients]
    return []

def generate_grocery_list(menu):
    """Generates a grocery list from the menu."""
    grocery_list = {}
    
    for day, meals in menu.items():
        for meal_type, recipe_str in meals.items():
            ingredients = extract_ingredients(recipe_str)
            for ingredient in ingredients:
                # Normalize ingredient to handle potential duplicates
                ingredient_normalized = ingredient.lower().strip()
                if ingredient_normalized in grocery_list:
                    grocery_list[ingredient_normalized] += 1
                else:
                    grocery_list[ingredient_normalized] = 1

    # Convert grocery list to a more readable format
    formatted_grocery_list = []
    for ingredient, count in grocery_list.items():
        if count > 1:
            formatted_grocery_list.append(f"{count} x {ingredient}")
        else:
            formatted_grocery_list.append(ingredient)
    
    return formatted_grocery_list


# Test the generate_grocery_list function
if __name__ == "__main__":
    grocery_list = generate_grocery_list(menu)
    print("\n".join(grocery_list))
```
